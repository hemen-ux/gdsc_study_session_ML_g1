## 1. Identify an ML Problem
Predicting whether a student will pass or fail an exam based on study hours and attendance.

## 2. Classify the ML Type
**Supervised Learning**

This is supervised learning because the model is trained using labeled data (pass or fail).
The algorithm learns the relationship between inputs (study hours, attendance) and the output (exam result).

## 3. Outline the ML Workflow
1. Collect data
2. Clean and preprocess data
3. Split data into training and testing sets
4. Train the model
5. Evaluate performance
6. Make predictions

## 4. One Possible Failure Point
**Noisy data**

Noisy or incorrect data can reduce model accuracy.
This can be fixed by cleaning the dataset and removing inconsistencies.
