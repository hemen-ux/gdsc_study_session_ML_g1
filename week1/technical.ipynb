```python
import numpy as np
```

```python
numbers=[3, 40, 5, 60, 10]
```

```python
array = np.array(numbers)
```

```python
Mean = np.mean(array)
Maximum = np.max(array)
Sum = np.sum(array)
```

```python
print("Numpy Array:", array)
print("Mean Value:", Mean)
print("Maximum Value:", Maximum )
print("Sum Value:" , Sum)
```
